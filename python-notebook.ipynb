## Data

For this analysis we'll use the `penguins` dataset from the [**palmerpenguins**](https://allisonhorst.github.io/palmerpenguins/) package.

```python
#| label: load-packages

import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")
```

## Species

Below is a scatterplot plot of species of penguins.

```python
sns.scatterplot(data=penguins, x="bill_length_mm",
  y="bill_depth_mm", hue="species", style="species", palette="colorblind")
plt.xlabel("Bill length (mm)")
plt.ylabel("Bill depth (mm)")
plt.show()
```
